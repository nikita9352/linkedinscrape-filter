```python
def get_infos():
    next_page = browser.find_element(By.XPATH, '//button[@aria-label="Next"]')
    time.sleep(3)
    browser.execute_script('arguments[0].scrollIntoView(true);', next_page)
    
    dict = {
        'name': [],
        'current_position': [],
        'current_company_duration': [],
        'current_company': [],
        'company_link' : [],
        'location': [],
        'url': [],

        'mckinsey_role': [],
        'mckinsey_time': [],
    }
    while len(dict['name'])<150:
        time.sleep(3)
        next_page = browser.find_element(By.XPATH, '//button[@aria-label="Next"]')
        browser.execute_script('arguments[0].scrollIntoView(true);', next_page)
        time.sleep(3)
        # document.body.scrollTop = document.body.scrollHeight;
        # document.documentElement.scrollTop = document.documentElement.scrollHeight;
        browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
        get_names = browser.find_elements(By.XPATH, '//a[@data-test-lead-result-entity="name"]')     

        get_current_company_title = browser.find_elements(By.XPATH, '//div[@data-test-lead-result-entity="title-at-company"]//span')
        get_current_company_nameandlink = browser.find_elements(By.XPATH, '//a[@data-test-lead-result-entity="current-company-link"]')
        get_timein_company = browser.find_elements(By.XPATH, '//div[@data-test-lead-result-entity="tenure"]')
        location = browser.find_elements(By.XPATH, '//div[@data-test-lead-result-entity="geo"]')
        get_Mckinseytime = browser.find_elements(By.XPATH, '//span[@data-test-result-experience="duration"]')
        get_Mckinsey_role  = browser.find_elements(By.XPATH, '//span[@data-test-result-experience="title"]')
        try:
            [dict['first_name'].append(get_names[i].text.split(' ')[0]) for i in range(len(get_names))]
            [dict['last_name'].append(get_names[i].text.split(' ')[-1]) for i in range(len(get_names))]
        except:
            dict['first_name'].append(None)
            dict['last_name'].append(None)
        try:
            [dict['current_company'].append(get_current_company_nameandlink[i].text) for i in range(len(get_current_company_nameandlink))]
        except:
            dict['current_company'].append(None)
        try:
            [dict['current_position'].append(get_current_company_title[i].text) for i in range(len(get_current_company_title))]
        except:
            dict['current_position'].append(None)
        try:
            [dict['current_company_duration'].append(get_timein_company[i].text) for i in range(len(get_timein_company))]
        except:
            dict['current_company_duration'].append(None)
        try:
            [dict['company_link'].append(get_current_company_nameandlink[i].get_property('href')) for i in range(len(get_current_company_nameandlink))]
        except:
            dict['company_link'].append(None)
        try:
            [dict['location'].append(location[i].text) for i in range(len(location))]
        except:
            dict['location'].append(None)
        try:
            [dict['mckinsey_role'].append(get_Mckinsey_role[i].text) for i in range(len(get_Mckinsey_role))]
        except:
            dict['mckinsey_role'].append(None)
        try:
            [dict['mckinsey_time'].append(get_Mckinseytime[i].text) for i in range(len(get_Mckinseytime))]
        except:
            dict['mckinsey_time'].append(None)
        time.sleep(3)
        next_page.click()
    return dict
```
